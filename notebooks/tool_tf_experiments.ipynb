```python
import numpy as np
from acrobotics.recources.torch_model import torch
from acrobotics.util import get_default_axes3d, tf_inverse, plot_reference_frame
from acrobotics.util import pose_y
```

```python
tf_path = pose_y(0.5, 0.0, 0.0, 0.1)
```

```python
fig, ax = get_default_axes3d(xlim=[-0.3, 0.3], ylim=[-0.3, 0.3], zlim=[-0.3, 0.3])
plot_reference_frame(ax, tf_path)
plot_reference_frame(ax, torch.tf_tt)
torch.plot(ax, c='k')
```

```python
tf_correction = np.dot(tf_path, tf_inverse(torch.tf_tt))
```

```python
fig, ax = get_default_axes3d(xlim=[-0.5, 0.5], ylim=[-0.5, 0.5], zlim=[-0.5, 0.5])
plot_reference_frame(ax, tf_path)
torch.plot(ax, tf=tf_correction, c='k')
```
