```python
import numpy as np
import matplotlib.pyplot as plt
from acrobotics.util import HaltonSampler, create_grid

np.random.seed(531)
```

```python
N = 10

a, b = np.linspace(-1, 1, N), np.linspace(-1, 1, N)
grid = create_grid([a, b])
grid.shape

rand = np.random.rand(N*N, 2)

hs = HaltonSampler(2)
halt = hs.get_samples(N*N)
```

```python
titles = ['Grid', 'Uniform random', 'Halton']
with plt.xkcd():
    fig, ax = plt.subplots(1, 3, figsize=(13, 4))
    ax[0].plot(grid[:, 0], grid[:, 1], 'x')
    ax[1].plot(rand[:, 0], rand[:, 1], 'x')
    ax[2].plot(halt[:, 0], halt[:, 1], 'x')

    for ai, title in zip(ax, titles):
        ai.set_xticks([])
        ai.set_yticks([])
        ai.set_title(title)
```
